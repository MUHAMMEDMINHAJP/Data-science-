```python
# importing matplotlib module
from matplotlib import pyplot as plt
from matplotlib import style
import  numpy as np

style.use('ggplot')
#x-axis values
x = [5,2,6,9,4,7,4]
#y-axis values
y = [10,5,7,6,4,8,9]
#function to plot
plt.plot(x,y,'o:c')
plt.title('Graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')
#function to show the plot
plt.show()

#subplot
x = np.array([0,3,5,7])
y = np.array([3,8,1,10])
plt.subplot(1,2,1)
plt.plot(x,y,color="C8",marker = 'o',mfc="r")
#plot 2:
x = np.array([0, 3, 4, 5, 2, 8, 3, 5, 8, 3, 7, 12, 13, 35])
y = np.array([10, 50, 60, 70, 80, 20, 6, 33, 87, 24, 63, 31, 52, 67])
plt.suptitle('subplot')
plt.subplot(1,2,2)
plt.scatter(x,y,marker = '*')
plt.show()

#histogram
plt.title('Histogram')
x = np.random.normal(60,15,150)
plt.hist(x,color="C2")
plt.show()

#piechart

plt.title('piechart')
y = np.array([15,25,55,51])
cars = ["Benz", "Honda", "Tata", "BMW"]
plt.pie(y, labels = cars)
plt.show()
```
